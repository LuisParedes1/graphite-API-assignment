## Dataset Preprocesing

For each document we
- Convert it to lowercase 
- Remove punctuation
- Perform Stemming

We don't
- Remove stop words because TfidfVectorizer does this
- Remove single quote (') because it might give us wrong words (don't -> dont)

```python
from nltk.stem import PorterStemmer

def removeSymbols(data:str):
    symbols = "!\"#$%&()*+-./:;<=>?@[\\]^_`{|}~\n\t“”—"
    
    for i in symbols:
        data = data.replace(i, ' ')
    
    return data

def lowerWords(data:str):
    return data.lower()

def stemWords(data:str):
    return PorterStemmer().stem(data)

def cleanDocument(document:str):
    document = lowerWords(document)
    document = removeSymbols(document)
    document = stemWords(document)
    return document
```

```python
train_dataset.loc[:,"title_and_content"] = train_dataset["title_and_content"].apply(cleanDocument)
display(train_dataset)
```

This is how a clean document looks now

```python
train_dataset.iloc[0]["title_and_content"]
```

# Training TF-IDF 

We train the TfidfVectorizer using the cleanned corpus

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = train_dataset["title_and_content"].to_list()
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words="english", lowercase=True, strip_accents='unicode')

# get tf-df values and train the model
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)
```

# Case Study: Getting the top TF-IDF terms for  [wikipedia tf-idf](https://en.wikipedia.org/wiki/Tf%E2%80%93idf)

As a case study, we scrape the [wikipedia tf-idf](https://en.wikipedia.org/wiki/Tf%E2%80%93idf) webpage's content and then obtain the terms with the highest TF-IDF

```python
import requests
from urllib.parse import unquote
from bs4 import BeautifulSoup
```

```python
url = "https%3A%2F%2Fen.wikipedia.org%2Fwiki%2FTf-idf"
response = requests.get(unquote(url))

parsed_website = BeautifulSoup(response.content, "html.parser")
```

```python
title = parsed_website.find('h1')

if title:
    title = title.get_text()
else:
    title = ""
```

```python
paragraphs = parsed_website.find_all('p')
all_text = ' '.join([p.get_text() for p in paragraphs])
all_text = title + ' ' + all_text
```

We do a bit of pre-processing for the text

```python
scraped_document = cleanDocument(all_text)
display(scraped_document)
```

## Getting the TF-IDF for [Wikipedia tf-idf](https://en.wikipedia.org/wiki/Tf%E2%80%93idf)

We transform the scrapped document and get it's tf-idf terms

```python
tfidf_url_page = tfidf_vectorizer.transform([scraped_document]).toarray()[0]
```

We order the TF-IDF terms in descending order and pick the top terms

```python
ordered_tfidf_url_page = sorted(zip(tfidf_vectorizer.get_feature_names_out(),
                                    tfidf_url_page), key=lambda x: x[1], reverse=True )
```

```python
limit = 10
ordered_tfidf_url_page[:limit]
```

We return the list in the expected format

```python
{ "terms": [
        {
            "term": term,
            "tf-idf": tfidf
        }
        for term, tfidf in ordered_tfidf_url_page[:limit]
    ]
}
```
